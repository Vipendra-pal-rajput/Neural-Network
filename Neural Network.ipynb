<h3>  Neural network form the base deep learning which is a sub field of machine learning.<br>
<br>
<font color="purple">What is preceptron : </font>
            A preceptron is a Neural network withot any hidden layer it only have input and output layer.
    
<font color="purple">where we can use preceptorn : </font>
    to give access to user or not for something.(like 0 & 1)
    
    A neural network has two steps : 
        1. Feedforward  : Gave random weight and find output.
        2. Backpropagation : compare error between predicted output or target output then update weight again using backprpagation.
    
<font color="red">Summarizing as ANN</font>
    
    1. Take input
    2. Add bias (if req.)
    3. Assign random weight to input features
    4. Run the code for training
    5. Find the error in prediction
    6. update the weight by gradient descent algorithm
    7. Repeat the training phase with updated weights
    8. Make predictions
    
</h3>  

<h3><font color='red'> example:</font>
    

        I/p - 0,1,2,3
        O/p - 0,2,4,6
    
        output = w1*i1 + w2*i2 + ------ + wn*in + b(bias)
        where : 
                wi = weight for input ii
                b = bias * weight 

        Now take w=3 :
                         I/p - 0,1,2,3
                  target O/p - 0,2,4,6
               Predicted o/p - 0,3,6,9
         error in prediction - 0,1,2,3 (Predicted o/p - target o/p)

    now we have two option Increase weight or decrease weight :
    
        Now take w=4 (Lets Increase weights) :
    
                         I/p - 0,1,2,3
                  target O/p - 0,2,4,6
               Predicted o/p - 0,4,8,12
         error in prediction - 0,2,4,6 (Predicted o/p - target o/p)
                now we can see error is also increasing
                
        Now take w=2 (lets decrease weight) :
                         I/p - 0,1,2,3
                  target O/p - 0,2,4,6
               Predicted o/p - 0,2,4,6
         error in prediction - 0,0,0,0 (Predicted o/p - target o/p)

     Now this w=2 is our final output because on this error is minimizes.
</h3>    

<h3>
    A sigmoid function serve as a activation function in neural N/W training. to get values in 0 & 1 we use sigmoid
    
    f(X) = 1/(1+e^-x)
    implementing OR gate using preceptron : 
    
        OR gate truth table:
                I/p - 0,0,1,1
                I/p - 0,1,0,1
                O/p - 0,1,1,1
    
    
    input for o1 = w1*x1 + w2*x2 + w3*x3
                   0.2*0 + 0.3*1 + 0.5*1
                   0.8
    
    b. calculate the output value :
        f(x) = 1/(1+e^-8)
             = 0.68997
    
    MSE = (1/2) * (target - output)^2
    
    MSE = (1/2) * (1-0.68997) = 0.048059
    
    Now we use back propagation
    

    
</h3>    

<h3>   
    Gradient Descent : 
        1. Start with initial values
        2. Calculate cost
        3. Update values using update function
        4. Return minimized cost for our function

        X = X-lr * (d/dX)f(X)
        
        X = input
        F(X) = output based on X
        lr = learning rate

        
        Derivation formula used in Neural N/W
            
            ∂error/∂w = ∂error/∂outo * ∂outo/∂ino  * ∂ino/∂w
            
            ∂error/ ∂outo = output - target
            
            ∂error/ ∂outo = (0.68997 - 1)= -0.31003
            
            ∂outo/∂ino = outo(1-outo)
                outo1 = (1/(1+e^-ino1))
                
          
            ∂ino/∂w = input values
            ∂ino/∂w = i2 = 1
            
            putting it all together :      -0.31003 * 0.21391 * 1 = -0.06631
                
                
            putting it in our main equation : 
               w = w - lr * ∂error/ ∂w2
        
        w2 = 1-(0.05)

```python
# giving Input values:
feature_set = np.array([[0,0],[0,1],[1,0],[1,1]])
print(feature_set.shape)
feature_set
```

```python
# our target output value
target_output = np.array([[0,1,1,1]])
target_output = target_output.reshape(4,1)
print(target_output.shape)
target_output
```

```python
# taking weights
weights=np.array([[0.1],[0.2]])
print(weights.shape)
weights
```

```python
# Adding Bias Values and Assigning a Learning Rate :
bias=0.3   # Bias weight
lr=0.05    # Learning rate
```

```python
def sigmoid(x):                  #defining a Sigmoid Function
    return 1/(1+np.exp(-x))    

def sigmoid_der(x):              # Derivative of sigmoid function
    return sigmoid(x)*(1-sigmoid(x))
```

```python
# predicting output and updating the weight values:

for epoch in range(20000):
    inputs = feature_set
    in_o = np.dot(feature_set, weights) + bias    # feedforward input
    out_o = sigmoid(in_o)                         #feedforward output
    error = out_o - target_output                 # backpropagation calculating error
    print(error.sum())                            # applying formula
    derror_douto = error                          # Calculating derivative
    douto_dino = sigmoid_der(out_o)
    deriv = derror_douto * douto_dino             # Multiplying individual derivatives
    inputs = feature_set.T                        # finding the transpose of input features
    deriv_final=np.dot(inputs, deriv)
    weights -= lr * deriv_final                   #updating the weight values
    for i in deriv:
        bias -= lr * i                            # updating the bias weight value
```

```python
print(weights)    # Final weight value
print(bias)       # Final bias value
```

```python
# Predictiong for (1,1)

in_put=np.array([1,1])    #taking input
result1= np.dot(in_put,weights)+bias   # 1st step
result2=sigmoid(result1)   # 2nd step
print("actual_result : ",result2)  # actual result
print("result_after_round : ",round(result2[0]))  #print result after round value
```

```python
# Predictiong for (1,0)

in_put=np.array([1,0])    #taking input
result1= np.dot(in_put,weights)+bias   # 1st step
result2=sigmoid(result1)   # 2nd step
print("actual_result : ",result2)  # actual result
print("result_after_round : ",round(result2[0]))  #print result after round value
```

```python
# Predictiong for (0,0)

in_put=np.array([0,0])    #taking input
result1= np.dot(in_put,weights)+bias   # 1st step
result2=sigmoid(result1)   # 2nd step
print("actual_result : ",result2)  # actual result
print("result_after_round : ",round(result2[0]))  #print result after round value
```
